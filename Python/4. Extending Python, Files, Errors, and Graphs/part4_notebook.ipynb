```python
#Exercise 66: Generating a Heatmap to Visualize the Grades of Students
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):
    if not ax:
        ax = plt.gca()

    im = ax.imshow(data, **kwargs)
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom")
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    ax.set_xticklabels(col_labels)
    ax.set_yticklabels(row_labels)
    ax.tick_params(top=True, bottom=False, labeltop=True, labelbottom=False)
    plt.setp(ax.get_xticklabels(), rotation=-30, ha="right", rotation_mode="anchor")
    for edge, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)
    return im, cbar
```

```python
#Exercise 66: Generating a Heatmap to Visualize the Grades of Students
import numpy as np
import matplotlib.pyplot as plt
data = np.array([
    [30, 20, 10,],
    [10, 40, 15],
    [12, 10, 20]
])
im, cbar = heatmap(data, ['Class-1', 'Class-2', 'Class-3'], ['A', 'B', 'C'], cmap='YlGn', cbarlabel='Number of Students')
```

```python
#Exercise 66: Generating a Heatmap to Visualize the Grades of Students
def annotate_heatmap(im, data=None, valfmt="{x:.2f}",
                     textcolors=["black", "white"],
                     threshold=None, **textkw):
    import matplotlib
    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)
    if isinstance(valfmt, str):
        valfmt = matplotlib.ticker.StrMethodFormatter(valfmt)
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            kw.update(color=textcolors[im.norm(data[i, j]) > threshold])
            text = im.axes.text(j, i, valfmt(data[i, j], None), **kw)
            texts.append(text)

    return texts
```

```python
#Exercise 66: Generating a Heatmap to Visualize the Grades of Students
im, cbar = heatmap(data, ['Class-1', 'Class-2', 'Class-3'], ['A', 'B', 'C'], cmap='YlGn', cbarlabel='Number of Students')
texts = annotate_heatmap(im, valfmt="{x}")
```

```python
# Exercise 69: Generating 3D plots to Plot a Sine Wave
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
X = np.linspace(0, 10, 50)
Y = np.linspace(0, 10, 50)
X, Y = np.meshgrid(X, Y)
Z = (np.sin(X))
# Setup axis
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
# Add title and axes labels
ax.set_title("Demo of 3D Plot", size=13)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
#Activity 13: Visualizing the Titanic Dataset Using a Pie Chart and Bar Plots
import pandas as pd
url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Python-Workshop/master/Chapter04/Activity13/titanic_train.csv'
titanic_df = pd.read_csv(url, delimiter=',')
print(titanic_df.head(5))
titanic_df.columns
```

```python
#Activity 13: Visualizing the Titanic Dataset Using a Pie Chart and Bar Plots
#Visualize the proportion of passengers that survived the incident (in a pie chart).
survived = titanic_df[titanic_df.Survived == 1].shape[0]
deceased = titanic_df[titanic_df.Survived == 0].shape[0]
print(survived, deceased)
# Pie chart.
labels = ['Survived', 'Deceased']
num = [survived, deceased]
import matplotlib.pyplot as plt
plt.pie(num, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'yellow'])
plt.title('Survived', fontdict={'fontsize': 20})
plt.pie(num, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'yellow'])
plt.show()
```

```python
#Activity 13: Visualizing the Titanic Dataset Using a Pie Chart and Bar Plots
#Compare the gender of passengers who survived the incident (in a bar plot).
survived_female = titanic_df[(titanic_df.Survived == 1) & (titanic_df.Sex == 'female')].shape[0]
survived_male = titanic_df[(titanic_df.Survived == 1) & (titanic_df.Sex == 'male')].shape[0]
# Plot bar chart.
sex = ['female', 'male']
num = [survived_female, survived_male]
plt.bar(sex, num)
plt.title('Surviving passengers count by gender')
plt.xlabel('Sex')
plt.ylabel('Num Survivors')
plt.show()
```
